```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constants
sigma = 5.67e-8  # Stefan-Boltzmann constant in W/(m^2*K^4)
A = 4e-4  # Area in m^2 (4 cm^2 converted to m^2)
BOAT_SURFACE_AREA = 0.005  # Surface area of the boat in m^2

# Temperatures in Kelvin
T1 = 900
T2 = 1800
```

```python
# Distance range in meters (from 30cm to 2cm)
r = np.linspace(0.02, 0.30, 500)

# Total power for each temperature
P_total_900K = sigma * T1**4 * BOAT_SURFACE_AREA
P_total_1800K = sigma * T2**4 * BOAT_SURFACE_AREA

# Incident power calculation
P_inc_900K = P_total_900K * (A / r**2)
P_inc_1800K = P_total_1800K * (A / r**2)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.semilogy(r, P_inc_900K, label='900K', color='b')
plt.semilogy(r, P_inc_1800K, label='1800K', color='r')
plt.xlabel('Distance (m)')
plt.ylabel('Incident Power (W)')
plt.title('Incident Power vs Distance')
plt.legend()
plt.grid(True)
plt.show()
```
